# Video Game Sales Analysis using Python

This project analyzes global video game sales data using Python.
The analysis explores sales by platform, genre, region, and publisher using data visualization techniques.

```python
import pandas as pd

df = pd.read_excel("Video game sales.xlsx")

df.head()
```

```python
df.shape
```

## Top Selling Video Games

```python
df.columns
```

```python
top_games = df.sort_values("global_sales", ascending=False)

top_games[["title","platform","global_sales"]].head(10)
```

## Sales by Platform

```python
import matplotlib.pyplot as plt

platform_sales = df.groupby("platform")["global_sales"].sum().sort_values(ascending=False).head(10)

platform_sales.plot(kind="bar")

plt.title("Top 10 Platforms by Global Sales")
plt.xlabel("Platform")
plt.ylabel("Global Sales")
plt.show()
```

```python
genre_sales = df.groupby("genre")["global_sales"].sum().sort_values(ascending=False)

genre_sales.plot(kind="bar")

plt.title("Global Sales by Game Genre")
plt.xlabel("Genre")
plt.ylabel("Global Sales")
plt.show()
```

```python
year_sales = df.groupby("release_year")["global_sales"].sum().sort_index()

year_sales.plot(kind="line")

plt.title("Global Video Game Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.show()
```

```python
publisher_sales = df.groupby("publisher")["global_sales"].sum().sort_values(ascending=False).head(10)

publisher_sales.plot(kind="bar")

plt.title("Top Publishers by Global Sales")
plt.xlabel("Publisher")
plt.ylabel("Global Sales")
plt.show()
```

## Correlation Between Sales Regions

```python
import seaborn as sns

corr = df[["north_american_sales",
           "european_sales",
           "japan_sales",
           "aisan_sales",
           "global_sales"]].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Between Sales Regions")

plt.show()
```

## Sales by Region

```python
region_sales = df[["north_american_sales", "european_sales", "japan_sales", "aisan_sales"]].sum()

region_sales.plot(kind="bar")

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
print("North America and Europe appear to contribute the most to global video game sales, while Japan and Asian sales are comparatively lower.")
```

## Conclusion

- Adventure and Shooter genres generate the highest global sales.
- PC and PlayStation platforms dominate the market.
